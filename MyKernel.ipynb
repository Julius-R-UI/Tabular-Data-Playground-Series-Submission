```python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.2)
```

```python
def objective(trial):
    params = {
        'reg_alpha' : trial.suggest_float('reg_alpha', 0, 10),
        'reg_lambda' : trial.suggest_float('reg_lambda', 0, 10),
        'num_leaves' : int(trial.suggest_int('num_leaves', 10, 200)),
        'learning_rate' : trial.suggest_loguniform('learning_rate', 0.001, 0.5),
        'n_estimators' : int(trial.suggest_int('n_estimators', 2000, 2200)),
        'min_child_samples': int(trial.suggest_int('min_child_samples', 1, 20)),
        'min_child_weight': trial.suggest_loguniform('min_child_weight', 0.001, 0.005),
        'subsample' : trial.suggest_float('subsample', 0.5, 1),
        'colsample_bytree' :trial.suggest_loguniform('colsample_bytree', 0.01, 0.09),
        }
    lgbm = LGBMClassifier(**params)
    lgbm.fit(x_train, y_train, eval_set=[(x_val, y_val)], early_stopping_rounds=200, verbose=False)
    return log_loss(y_val,lgb.predict_proba(x_val))

optuna.logging.set_verbosity(optuna.logging.WARNING)
study = optuna.create_study(direction='minimize')
study.optimize(objective)

params_for_lgbm = study.best_trial.params
```
